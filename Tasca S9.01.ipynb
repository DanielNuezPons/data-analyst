1. Calculadora de l'√≠ndex de massa corporal

Escriu una funci√≥ que calculi l'IMC ingressat per l'usuari/√†ria, √©s a dir, qui ho executi haur√† d'ingressar aquestes dades. Pots obtenir m√©s informaci√≥ del seu c√†lcul en: üîó √çndice de masa corporal IMC que es y como se calcula.

La funci√≥ ha de classificar el resultat en les seves respectives categories.

Consell: Intenta validar les dades pr√®viament, perqu√® envi√Ø un missatge d‚Äôadvert√®ncia si les dades introdu√Ødes per l‚Äôusuari/a no estan en el format adequat o no pren valors raonables.

```python
def clasificaci√≥n_imc():
    try:
        # solicitamos los datos al usuario
        peso = float(input('Ingrese su peso en kg: '))
        altura = float(input('Introduce tu altura en m: '))

        # validaci√≥n de los datos introducidos
        if peso <= 0 or peso >= 300:
            raise ValueError('Error el peso tiene que ser positivo e inferior a 300 kg. ')
            
        if altura <= 0 or altura >= 2.5:
            raise ValueError('Error la altura tiene que ser positiva e inferior a 2.5m ')
            

        # Calculamos el IMC
        imc = peso / (altura**2)

        # clasificacion segun el IMC
        if imc < 18.5:
            print(f"Tu IMC es {round(imc, 2)}, Categor√≠a: Bajo peso")
        elif 18.5 <= imc < 24.9:
            print(f"Tu IMC es {round(imc, 2)}, Categor√≠a: Peso normal")
        elif 25 <= imc < 29.9:
            print(f"Tu IMC es {round(imc, 2)}, Categor√≠a: Sobrepeso")
        else:
            print(f"Tu IMC es {round(imc, 2)}, Categor√≠a: Obesidad")

    except ValueError as ValorError:
        print("Por favor, ingrese valores num√©ricos v√°lidos para el peso y la altura.\n", ValorError)



  


```

2. Convertidor de temperatures

Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les m√©s comunes s√≥n Celsius (¬∞C), Fahrenheit (¬∞F) i Kelvin (K). Tamb√© existeixen altres unitats com Rankine (¬∞Ra) i R√©aumur (¬∞Re).

Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a m√≠nim, dues conversions, de manera que es puguin guardar (recorda que un print() no es pot guardar mai).

Consell: Intenta validar les dades pr√®viament, perqu√® envi√Ø un missatge d‚Äôadvert√®ncia si les dades introduides per l‚Äôusuari/a no estan en el format adequat.

(EXTRA): Pensa una manera d‚Äôemmagatzemar totes les possibles conversions en un sol objecte (Llista? Diccionari? DataFrame?) en comptes d‚Äôescriure molts if else en funci√≥ de la temperatura d‚Äôorigen i la temperatura de dest√≠.

```python
# Definimos la funci√≥n
def convertir_temperatura():
    try:
        valor = float(input('Introduce el valor de la temperatura: '))
        unidad_origen = input('Introduce la unidad (C, F, K): ').strip().upper()

        # Validaci√≥n unidad de origen
        if unidad_origen not in ['C', 'F', 'K']:
            raise ValueError('Unidad de origen no existente. Utiliza C, F, K. ')

        # Creamos un nuevo diccionario para guardar las conversiones 
        conversiones_temp = {}

        # conversi√≥n desde Celsius 
        if unidad_origen == 'C':
            conversiones_temp['Celsius'] = valor
            conversiones_temp['Fahrenheit'] = (valor * 9/5) + 32
            conversiones_temp['Kelvin'] = valor + 273.15
            conversiones_temp['Rankine'] = (valor + 273.15) * 9/5
            conversiones_temp['R√©aumur'] = valor * 0.8

        # conversi√≥n desde Fahrenheit
        elif unidad_origen == 'F':
            celsius = (valor - 32) * 5/9
            conversiones_temp['Celsius'] = celsius
            conversiones_temp['Fahrenheit'] = valor
            conversiones_temp['Kelvin'] = celsius + 273.15
            conversiones_temp['Rankine'] = valor + 459.67
            conversiones_temp['R√©aumur'] = celsius * 0.8

        # conversi√≥n desde Kelvin
        elif unidad_origen == 'K':
            celsius = valor - 273.15
            conversiones_temp['Celsius'] = celsius
            conversiones_temp['Fahrenheit'] = (celsius * 9/5) + 32
            conversiones_temp['Kelvin'] = valor
            conversiones_temp['Rankine'] = valor * 1.8
            conversiones_temp['R√©aumur'] = celsius * 0.8

        # mostrar los resultados

        print("\nConversi√≥n actual:")
        for unidad, resultado in conversiones_temp.items():
            print(f"{unidad}: {round(resultado, 2)}")
        
        for unidad in conversiones_temp:
            conversiones_temp[unidad] = round(conversiones_temp[unidad], 2)
        print(conversiones_temp)

    except ValueError as valorError:
        print('Ingrese valor num√©rico o unidad correcta.\n', valorError)


        

            



```

```python
# Llamar a la funci√≥n
convertir_temperatura()
```

3. Comptador de paraules d'un text.

Escriu una funci√≥ que donat un text, mostri les vegades que apareix cada paraula. Intenta que es gestionin totes les casu√≠stiques possibles que facin que el programa no funcioni correctament.

(EXTRA): Quina √©s la llargada mitjana de les paraules del text que has escrit? ‚ÄúHola com va?‚Äù hauria de retornar (4+3+2) / 3 = 3

```python
import string
def contador_palabras():
    try:
        texto = input('introduce un texto: ')
        
        # validacion de texto 
        if not texto.split():
            raise ValueError('Introduce un texto que no este vacio')
        
        # pasar todo el texto a minusculas
        texto = texto.lower()

        # Eliminaci√≥n de todos los signos estandar
        traductor = str.maketrans('', '', string.punctuation)
        texto = texto.translate(traductor)

        # separaci√≥n de las palabras
        palabras = texto.split()

        # Validaci√≥n de palabras en el texto
        if not palabras:
            raise ValueError ('No hay palabras en el texto: vuelve a escribirlo')
        
        # Realizamos el conteo de las palabras creando un diccionario
        contador = {}
        for p in palabras:
            if p in contador:
                contador[p]  += 1 
            else:
                contador[p] = 1
        
        # Calculo de la longitud y longitud media de las palabras
        longitud = [len(p) for p in palabras]
        longitud_media = sum(longitud) / len(longitud)

        # Mostrar los resultados 
        print('\nConteo de palabras:')
        for palabra, cantidad in contador.items():
            print(f'{palabra} : {cantidad} ')
        
        print(f'\nNumero total de palabras : {len(palabras)}')
        print(f'Longitud media de las palabras : {round(longitud_media, 2)}')
    
    except ValueError as valorError:
        print('Ingrese un texto con palabras.\n', valorError)

```

```python
# llamamos a la funci√≥n
contador_palabras()
```

1. Comptador i endre√ßador de paraules d'un text.

El client va quedar content amb el comptador de paraules, per√≤ ara vol llegir arxius TXT i que calculi la freq√º√®ncia de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, √©s a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z. Per exemple, per a l'arxiu "tu_me_quieres_blanca.txt" la sortida esperada seria:

```python
import string
from pprint import pprint

# Definimos la funci√≥n
def contador_palabras_txt(file_path):
    try:
        #  Abrir y leer el archivo TXT
        with open(file_path, "r", encoding="utf-8") as f:
            texto = f.read()

        #  Validaci√≥n de archivo o texto vacio
        if not texto.strip():
            raise ValueError("El archivo est√° vac√≠o. Introduce un texto v√°lido.")

        # Pasar todo el archivo a min√∫sculas
        texto = texto.lower()

        # Eliminaci√≥n de todos los signos estandar
        traductor = str.maketrans('', '', string.punctuation)
        texto = texto.translate(traductor)

        # Separar las palabras
        palabras = texto.split()

        if not palabras:
            raise ValueError("No se encontraron palabras v√°lidas en el archivo.")

        # Crear diccionario  A-Z
        diccionario_final = {letra: {} for letra in string.ascii_lowercase}

        # Variable para calcular la media de longitud del archivo
        total_letras = 0

        # Recorrer cada palabra y contar frecuencia
        for palabra in palabras:
            primera_letra = palabra[0]

            # Si la palabra empieza con un car√°cter no alfab√©tico, la saltamos
            if primera_letra not in diccionario_final:
                continue
            diccionario_final[primera_letra][palabra] = diccionario_final[primera_letra].get(palabra, 0) + 1
            total_letras += len(palabra)

        # Ordenar letras y palabras internamente
        diccionario_final = dict(sorted(diccionario_final.items()))
        diccionario_final = {k: dict(sorted(v.items())) for k, v in diccionario_final.items()}

        # Imprimir resultado como diccionario
        pprint(diccionario_final)


        # Calcular media de longitud de palabras
        media_longitud = total_letras / len(palabras)
        print(f"\nN√∫mero total de palabras: {len(palabras)}")
        print(f"Longitud media de las palabras: {round(media_longitud, 1)}")

    except FileNotFoundError:
        print(" Error: No se encontr√≥ el archivo. Verifica la ruta.")
    
```

2. Conversi√≥ de tipus de dades.

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client:

```python
lista_cliente = [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

def separar_flotantes(lista):
    lista_float = []
    lista_no_float =[]

    for elem in lista:
        if isinstance (elem,(list, tuple)):
            conv, no_conv = separar_flotantes(elem)
            lista_float.extend(conv)
            lista_no_float.extend(no_conv)

        else:
            try:
                lista_float.append(float(elem))
            except (ValueError, TypeError):
                lista_no_float.append(elem)

    return lista_float, lista_no_float
```

Explora el funcionament del m√≤dul random de la llibreria numpy. üîó Random module in NumPy

En aquest punt, el client ha detectat un problema amb les contrasenyes que utilitzen els seus treballadors. Asdf1234, dates d‚Äôaniversari o similars. Per a solucionar-ho ens ha encarregat una funci√≥ de Python que generi contrasenyes m√©s segures. La funci√≥ ha de dependre dels seg√ºents par√†metres:

longitud (int): Longitud de la contrasenya
majuscules (bool = True): Si hi ha d‚Äôapar√®ixer maj√∫scules
minuscules (bool = True): Si hi ha d‚Äôapar√®ixer min√∫scules
numeros (bool = True): Si hi ha d‚Äôapar√®ixer n√∫meros
signes (bool = False): Si hi ha d‚Äôapar√®ixer car√†cters especials (,-$? o similars)
Aix√≠ doncs, si executem la funci√≥ de la seg√ºent manera:

crear_contrasenya(10, True, True, True, True)

Haur√≠em d‚Äôobtenir un output (que hem de poder guardar) de l‚Äôestil:

9Er,5Vn8P$

Assegura‚Äôt que es compleixin tots els criteris, i que aquestes contrasenyes siguin realment aleat√≤ries.

(EXTRA) Explora com podr√≠em fer que la funci√≥ copi√©s la contrasenya autom√†ticament al porta papers de l‚Äôordinador (com si l‚Äôhagu√©ssim seleccionat i fet ctrl+copy).

```python
import string
import numpy as np
import random
import pyperclip

# definimos la funci√≥n y el tipo de caracteres que vamos a usar
def crear_contrase√±a(longitud=10, mayusculas=True, minusculas=True, numeros=True, signos=False):
    caracteres = ""
    if mayusculas:
        caracteres += string.ascii_uppercase
    if minusculas:
        caracteres += string.ascii_lowercase
    if numeros:
        caracteres+= string.digits
    if signos:
        caracteres += string.punctuation

    if not caracteres:
        return "Debe seleccionar al menos un tipo de car√°cter"
    
    # Generamos la contrase√±a 

    contrase√±a = ''.join(random.choices(caracteres, k=longitud))

    # Copiar la contrase√±a al portapapeles
    pyperclip.copy(contrase√±a)
    return(contrase√±a) 
    
 
```

2. Processament de dades simple

Una companya de feina ens ha demanat un favor, aprofitant que sap que estem aprenent a programar. T√© un hist√≤ric de partits de futbol catal√† en un fitxer, on hi ha emmagatzemat els noms dels equips i els resultats. Necessita que processem les dades de manera autom√†tica, per extreure‚Äôn els resultats que necessita.

Utilitza el fitxer "historic_partits.txt"

Necessita un programa que retorni:

El nombre total de gols que ha fet cada equip.
El nom de l‚Äôequip m√©s golejador.
El nom de l‚Äôequip m√©s golejat
La classificaci√≥ global (cada vict√≤ria: 3 pts, empat 1 pts, derrota 0 pts)
Com a refer√®ncia, els equips amb m√©s punts serien els seg√ºents:

imatge resultat partits
Pots utilitzar el format que et sigui m√©s convenient.

```python
# definimos la funci√≥n
def analisis_resultados_txt(file_path):
    goles_hechos = {}
    goles_recibidos = {} 
    puntos = {}

    with open(file_path, 'r', encoding='utf-8') as archivo:
        for linea in archivo:
            partes = linea.strip().split('\t')
            if len(partes) != 3:
                continue
        
            equipo_local, resultado, equipo_visitante = partes
            try:
                goles_local, goles_visitante = map(int, resultado.split('-'))
            except ValueError:
                continue

            # Goles hechos
            goles_hechos[equipo_local] = goles_hechos.get(equipo_local, 0) + goles_local
            goles_hechos[equipo_visitante] = goles_hechos.get(equipo_visitante, 0) + goles_visitante

            # Goles recibidos
            goles_recibidos[equipo_local] = goles_recibidos.get(equipo_local, 0) + goles_visitante
            goles_recibidos[equipo_visitante] = goles_recibidos.get(equipo_visitante, 0) + goles_local

            # Inicializamos los puntos
            if equipo_local not in puntos:
                puntos[equipo_local] = 0
            if equipo_visitante not in puntos:
                puntos[equipo_visitante] = 0

            # Clasificaci√≥n por puntos
            if goles_local > goles_visitante:
                puntos[equipo_local] += 3
            elif goles_local < goles_visitante:
                puntos[equipo_visitante] += 3 
            else:
                puntos[equipo_local] += 1
                puntos[equipo_visitante] += 1
    
    equipo_goleador = max(goles_hechos.items(), key=lambda x: x[1])
    equipo_goleado = max(goles_recibidos.items(), key=lambda x: x[1])

    return goles_hechos, equipo_goleador, equipo_goleado, puntos
```

```python
goles_por_equipo, goleador, goleado, clasificacion = analisis_resultados_txt(r"C:\Users\nuez_\Desktop\BOOTCAMP DATA\historic partits.txt")

print("Goles marcados por cada equipo:")
for equipo, goles in sorted(goles_por_equipo.items(), key=lambda x: x[1], reverse=True):
    print(f"{equipo}: {goles} goles")

print(f"\nEquipo m√°s goleador: {goleador[0]} con {goleador[1]} goles")
print(f"Equipo m√°s goleado: {goleado[0]} con {goleado[1]} goles recibidos")

```
