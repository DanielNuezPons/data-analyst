6.
Crea una columna nova que sigui la data de naixament en format Datetime a partir de les columnes dia, mes i any. Utilitzant aquesta columna crea una funció que donada una data, et calculi l'edat actual a dia d'avui.

Utilitza la funció que acabes de crear per generar una columna nova al DataFrame amb l'edat actual.

```python
from datetime import datetime, date

# Creamos la columna any de naixement 
df["Data de naixement"] = pd.to_datetime({'year':df["Any"], 'month':df["Mes"], 'day':df["Dia"]})

# definimos la función para obtener la edad
def edad (naixement):
    today=date.today()
    return today.year - naixement.year - ((today.month, today.day) < (naixement.month, naixement.day ))

df["Edad actual"] = df["Data de naixement"].apply(edad)
df
```

La funció ha de crear (i exportar) un gràfic automàticament per a cada columna del dataframe. Per exemple:

un histograma/boxplot si la variable és numèrica
unes barres dels valors més freqüents si és categòrica

La idea és crear una funció que funcioni per qualsevol dataframe, no només amb el que hem treballat fins ara.

Mostra el resultat de la funció en algun dels datasets d’exemple que conté el paquet seaborn. Per exemple, iris, penguins o titanic.

Tingues en consideració que en el següent sprint treballaràs exclusivament amb gràfics. L’objectiu d’aquest exercici no és crear gràfics molt elaborats, sinó resoldre una necessitat de manera ràpida i automàtica.

```python
import matplotlib.pyplot as plt


# Definimos la función 
def columna_gráfico(df):
    for col in df.columns: # recorremos todas las columnas del Dataframe
        if df[col].dtype in ['int64', 'float64']:  # si la columna es tipo entero o float
            plt.figure()
            df[col].plot.hist(edgecolor='black', color='green')
            plt.title(col)
            plt.xlabel(col)
            plt.ylabel("frecuencia")
            plt.tight_layout()
            plt.savefig(f"grafico_numerico_{col}.png")
            plt.show()
            plt.close()
            print(f"Grafico de la columna {col} guardado como:\ngrafico_numerico_{col}.png")

        elif df[col].dtype in ['object']: # Si la columna es tipo categórico
            plt.figure()
            df[col].value_counts().plot.bar(edgecolor='black',color='skyblue')
            plt.title(col)
            plt.xlabel(col)
            plt.ylabel('frecuencia')
            plt.tight_layout()
            plt.savefig(f"grafico_categorico_{col}.png")
            plt.show()
            plt.close()
            print(f"Grafico de la columna {col} guardado como:\ngrafico_categorico_{col}.png")

        elif df[col].dtype in ['datetime64[ns]']: # si la columna es tipo datetime
            plt.figure()
            df[col].value_counts().sort_index().plot.bar(edgecolor='black', color='orange')
            plt.title(col)
            plt.xlabel(col)
            plt.ylabel('frecuencia')
            plt.tight_layout()
            plt.savefig(f"grafico_datetime_{col}.png")
            plt.show()
            plt.close()
            print(f"-Grafico de la columna {col} guardado como:\ngrafico_datetime_{col}")


import seaborn as sns
penguins = sns.load_dataset('penguins')
columna_gráfico(penguins)

```

Carrega l'arxiu matriu_distancies.xlsx a pandas, de manera que els noms de files i els noms de columnes siguin els de les ciutats. Borra "Las Palmas de Gran Canaria" i "Palma" perquè poguem fer el trajecte en cotxe.

Font: Mejores Rutas

Ens interessa visitar totes les ciutats principals d'Espanya recorrent la mínima distància possible.

No cal que ho facis de forma òptima, ens interessa que desenvolupis una solució raonable utilitzant les eines que tens actualment.

Per exemple, una aproximació senzilla (que no òptima) seria anant sempre a la ciutat més propera que no haguem visitat encara

Fes una funció que donada la matriu de distàncies i la ciutat d'origen, faci una proposta de ruta que sigui el més curta possible que puguis, retornant una llista amb l'ordre de visita. Dóna també la distància total recorreguda.

(EXTRA) Des de quina ciutat la ruta seria més curta amb l'algoritme plantejat

```python
import pandas as pd
import numpy as np
ruta_df = pd.read_excel(r"C:\Users\nuez_\Downloads\matriu_distancies.xlsx", index_col=0) # Cargamos el DataFrame

# Eliminamos la palma y las palmas de gran canaria del DataFrame

ruta_df= ruta_df.drop(index=['Palma','Las Palmas de Gran Canaria'], columns=['Palma','Las Palmas de Gran Canaria'])

ruta_df
```

```python
# definimos la función
def ruta_mas_corta(ruta_df, origen):
    
    # creamos una lista con la ciudad de origen como primera ciudad visitada, contador para la distancia y ponemos como c_actual el origen de la ruta
    c_visitadas=[origen]
    distancia_total=0
    c_actual=origen

    #Generamos un bucle que recorrera todas las ciudades del Dataframe

    while len(c_visitadas) < len(ruta_df):
        distancia= ruta_df.loc[c_actual].drop(c_visitadas) # obtenemos la distancia entre la ciudad actual y las no visitadas 
        siguiente= distancia.idxmin() # Calculamos la distancia mas corta a recorer
        distancia_total += distancia[siguiente] # sumamos la distancia recorrida
        # actualizamos la ruta y la ciudad actual
        c_visitadas.append(siguiente)
        c_actual= siguiente
    return c_visitadas, distancia_total

    


```

```python
ruta, distancia = ruta_mas_corta(ruta_df, "Barcelona")
print("Ruta propuesta:", " -> ".join(ruta))
print("Distancia total:", distancia, "km")

```
